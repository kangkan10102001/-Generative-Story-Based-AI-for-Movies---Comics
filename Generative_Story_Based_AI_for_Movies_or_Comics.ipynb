```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can replace it with "EleutherAI/gpt-neo-1.3B" for GPT-Neo
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Encode the prompt and generate text
def generate_story(prompt):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=500, num_return_sequences=1, no_repeat_ngram_size=2)
    story = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return story

# Example usage
story = generate_story("Write a plot for a futuristic sci-fi movie about AI.")
print(story)
```

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2", torch_dtype=torch.float32)

pipe.to("cpu")

# Generate an image from a prompt
def generate_image(prompt):
    image = pipe(prompt).images[0]
    image.show()
    return image

# Example usage with a scene description
image = generate_image("A futuristic city with flying cars and neon lights.")

```

```python
import gradio as gr

# Define the function to return story and image
def generate_story_and_image(prompt):
    story = generate_story(prompt)
    image = generate_image(story)  # Use the generated story as input for image creation
    return story, image

# Create the Gradio interface
iface = gr.Interface(fn=generate_story_and_image,
                     inputs="text",
                     outputs=["text", "image"])

iface.launch()
```
